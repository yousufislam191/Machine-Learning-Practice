# Seaborn


*   Data Visualization Library



```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: Seaborn has some built-in dataset

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

### Scatter Plot

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
sns.scatterplot(x='sepal_width', y='petal_width', hue='species', data=iris)
```

```python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

### Count Plot

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(x='survived', data=titanic)
```

```python
sns.countplot(x='age', data=titanic)
```

### Bar Chart

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
# cancer dataset
from sklearn.datasets import load_breast_cancer
breast_cancer_dataset = load_breast_cancer()
print(breast_cancer_dataset)
```

```python
# set breast cancer datasate to pandas DataFrame
breast_cancer_df = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
breast_cancer_df.head()
```

```python
# add a column
breast_cancer_df['breast cancer detected'] = breast_cancer_dataset.target
breast_cancer_df.head()
```

### Distribution Plot

```python
sns.displot(breast_cancer_df['mean texture'])
```

### Correlation


1.   Positive Correlation : when **x** value increases at the same time **y** value increases.
2.   Negative Correlation : when x value increases at the same time y value decreases.





```python
correlation = breast_cancer_df.corr()
correlation.head()
```

```python
# constructing a Heat Map
plt.figure(figsize=(20, 15))
sns.heatmap(correlation, cbar=True, square=True, fmt='.2f', annot=True, annot_kws={'size':10}, cmap='Blues')
```
